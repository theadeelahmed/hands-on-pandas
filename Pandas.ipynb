##  Merge, Join and Concatenate

```python
data1 = {"Emp_id":["E01","E02","E03","E04","E05","E06"],
         "Name":["Adeel","Ali","Aliza","Asghar","Sarah","Ahmed"],
         "Age" :[22,34,21,31,25,29]}

data2 = {"Emp_id" : ["E01","E02","E03","E08","E05","E07"],
        "Salary": [80000,65000,82000,30000,38000,43000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(df1)

print(df2)
```

```python
pd.merge(df1,df2, on= "Emp_id")     #merging behave of emp_id column
```

```python
pd.merge( left = df1, right = df2, on= "Emp_id", how = "left")   #left join show All left data and common right data 
```

```python
pd.merge(left= df1, right= df2, on= "Emp_id", how = "right")   # show right side full data 
```

```python
df1 = pd.DataFrame({"Emp_id":["E01","E02","E03","E04","E05","E06"],
         "Name":["Adeel","Ali","Aliza","Asghar","Sarah","Ahmed"],
         "Age" :[22,34,21,31,25,29]})

df2 = pd.DataFrame({"Emp_id":["E07","E08","E09","E10","E11","E12"],
         "Name":["Kabeer","Nasir","Amber","Owais","Hashir","David"],
         "Age" :[22,34,21,31,25,29]})
```

```python
# Concatenate
pd.concat([df1,df2])
```

##  Compare DataFrame 

```python
df1 = pd.DataFrame({"Fruits":["mango","apples","bananas","papaya"],
                    "Price":[100,180,80,80,],
                    "Quantity":[15,10,10,3]})

df2 = df1.copy()
```

```python
df2.loc[0,"Price"]=120           #updating/changing price and Quantity
df2.loc[1,"Price"]=200
df2.loc[3,"Price"]=65

df2.loc[0,"Quantity"]=12
df2.loc[1,"Quantity"]=15
df2.loc[3,"Quantity"]=5

df2
```

```python
df1.compare(df2)   # comparing new and old price and Quantity
```

```python
df1.compare(df2,align_axis=0)
```

```python
df1.compare(df2,keep_shape=True)    #show NaN in not change values 
```

# Pivoting and  melting  DataFrames

```python
dict =   {"Keys":["k1","k2","k1","k2"],
          "Names":["Adeel","chirs","Raza","john"],
          "Houses":["Red","Blue","Green","Red"],
          "Grades":["3rd","8th","9th","8th"]}

data = pd.DataFrame(dict)
```

```python
data.pivot(index="Keys", columns="Names", values=["Houses","Grades"])
```

```python
#melting
   
df =  pd.DataFrame({"Names":["Adeel","chirs","Raza","john"],
          "Houses":["Red","Blue","Green","Red"],
          "Grades":["3rd","8th","9th","8th"]})
```

```python
df
```

```python
pd.melt(df, id_vars=["Names"],value_vars=["Houses","Grades"])  #melting and creating new columns 
```

```python
pd.melt(df, id_vars=["Names"],value_vars=["Houses","Grades"],var_name="House&Grades",value_name="values")   # changing variable name  
```
